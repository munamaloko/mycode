```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange', 'pink', 'banana']
counts = [40, 50, 60, 70, 60, 50]
bar_labels = ['red', 'blue', '_red', 'orange', 'pink', 'yellow']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange', 'tab:pink', 'tab:yellow']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```
